# Ensembles - Bagging and Forests



Continuamos nossas aulas de técnicas de *machine learning*. Até agora vimos como montar modelos isolados para tarefas de classificação e regrssão. Cada modelo, por seu funcionamento e capacidade, acaba tendo seus prós e contras: alguns focam em certas métricas, outros encaram o trade-off de viés e variância de certo modo. Dessa forma, escolher um algoritmo acarreta diretamente no comportamento e trade-offs do seu modelo. Mas e se não usássemos apenas um, mas vários modelos que se completam. Enter Ensembles! Nas próximas aulas, veremos como encarar problemas utilizando técnicas de Ensemble, que nada mais são do que juntar modelos diversos para termos uma melhor modelagem.

O *Ensemble Learning* ajuda a melhorar os resultados do *Machine Learning* combinando vários modelos. Essa abordagem permite a produção de melhor desempenho preditivo em comparação com um único modelo. É por isso que os métodos de ensemble são os primeiros em muitas competições de aprendizado de máquina de prestígio, como a competição Netflix, KDD 2009 e Kaggle.

Os métodos de *Ensemble* são meta-algoritmos que combinam várias técnicas de aprendizado de máquina em um modelo preditivo para diminuir a variância (*Bagging*), o viés (*Boosting*) ou melhorar as predições (*Stacking*).



Os métodos de *Ensembles* podem ser divididos em dois grupos:


1. ***Sequential Ensemble Methods***: métodos de conjunto sequencial em que os *base learners* são gerados sequencialmente (exemplos são os métidos de *Boosting*). A motivação básica dos métodos sequenciais é explorar a dependência entre os *base learners*. O desempenho geral pode ser aprimorado pesando exemplos previamente rotulados incorretamente com maior peso.


2. ***Parallel Ensemble Methods***: métodos de conjunto paralelo em que os *base learners* são gerados em paralelo (exemplos são os métodos de *Bagging, Stacking, Voting*). A motivação básica dos métodos paralelos é explorar a independência entre os *base learners*, pois o erro pode ser reduzido drasticamente pela média.


A maioria dos métodos de *Ensemble* usa um algoritmo de aprendizado de base única para produzir *base learners* homogêneos, ou seja, alunos do mesmo tipo, levando a *Ensembles* homogêneos.

Existem também alguns métodos que usam *base learners* heterogêneos, ou seja, *learners* de diferentes tipos, levando a *Ensembles* heterogêneos. Para que os métodos de *Ensemble* sejam mais precisos do que qualquer um de seus membros individuais, os *base learnenrs* precisam ser o mais precisos possível e o mais diversificados possível.

Os métodos de ensemble são usados em quase todos os hackathons de ML para aprimorar as habilidades de previsão dos modelos. Vamos dar uma olhada nas vantagens de usar métodos de conjunto:

# Voting Ensemble




O EnsembleVoteClassifier é um meta-classificador para combinar classificadores de aprendizado de máquina semelhantes ou conceitualmente diferentes para classificação por votação por maioria ou por pluralidade. (Para simplificar, vamos nos referir à votação por maioria e pluralidade como votação por maioria.)

O EnsembleVoteClassifier implementa votação "hard" e "soft". Na votação definitiva, prevemos o rótulo da classe final como o rótulo da classe que foi predito com mais frequência pelos modelos de classificação. Na votação *soft*, prevemos os rótulos das classes calculando alguma estatística com as probabilidades de classe (recomendadas apenas se os classificadores estiverem bem calibrados).

Se você estiver interessado em usar o EnsembleVoteClassifier, observe que agora ele também está disponível no scikit learn (> 0,17) como VotingClassifier.




Você pode treinar seu modelo usando diversos algoritmos e agrupá-los para predizer a saída final. Digamos, você usa um *Random Forrest Classifier*, *SVM Classifier*, *Logistic Regression*, etc .; os modelos são colocados um contra o outro e selecionados com o melhor desempenho votando usando a classe VotingClassifier de sklearn.ensemble.

É o jeito mais simples de se fazer um Ensemble - construa vários algoritmos sob o mesmo dataset de treino e deixe que cada um dê seu output. Com o output de todos em mãos, podemos:

 - ***Majority Vote / Hard Voting***
 

 -  ***Weighted Majority Voting***
 
 
 - ***Soft Voting***  

## Majority Voting / Hard Voting

Na votação majoritária/forçada, temos um conjunto de classificadores, heterogêneos ou homogêneos, dos quais recolheremos as previsões.

$$ \hat{y} = mode [ C_1(x), C_2(x), ..., C_n(x) ]  $$

Supondo que combinemos três classificadores que classificam uma amostra de treinamento da seguinte maneira:

classificador 1 -> classe 0
classificador 2 -> classe 0
classificador 3 -> classe 1

$$ \hat{y} = mode [ 0, 0, 1 ] = 0  $$

Por maioria de votos, classificaríamos a amostra como "classe 0."

## Weighted Majority Vote

Além do voto majoritário simples (votação forçada), conforme descrito na seção anterior, podemos calcular um voto majoritário ponderado associando um peso $w_j$ ao classificador $C_j$.


$$ \hat{y} = argmax_i \sum_{j=1}^m w_j X_a (C_j(x) = i)  $$

onde $X_a$ é a função característica $[C_j (x) = i∈A]$ e A é o conjunto de rótulos de classe exclusivos.

Continuando com o exemplo da seção anterior

- classificador 1 -> classe 0
- classificador 2 -> classe 0
- classificador 3 -> classe 1

atribuir os pesos {0.2, 0.2, 0.6} produziria uma previsão $\hat{y} = 1$:

$$ \hat{y} = argmax_i [ 0.2  i_0 + 0.2  i_0 + 0.6  i_1 ] = 1 $$

## Soft Voting

Na votação *soft*, prevemos os rótulos das classes com base nas probabilidades previstas p para o classificador - essa abordagem é recomendada apenas se os classificadores estiverem bem calibrados.

$$ \hat{y} = argmax_i \sum_{j=1}^m w_i p_{ij}  $$

onde $w_j$ é o peso que pode ser atribuído ao j-ésimo classificador.

Supondo que o exemplo na seção anterior fosse uma tarefa de classificação binária com rótulos de classe i∈ {0,1}, nosso conjunto poderia fazer a seguinte previsão:

- C1 (x) → [0.9, 0.1]
- C2 (x) → [0.8, 0.2]
- C3 (x) → [0.4, 0.6]

Usando pesos uniformes, calculamos as probabilidades médias:

$$ p(i_0 | X) = \frac {0.9 + 0.8 + 0.4} {3} = 0.7  $$
<br>
<br>
$$ p(i_1 | X) = \frac {0.1 + 0.2 + 0.6} {3} = 0.3  $$
<br>
<br>
$$ \hat{y} = argmax_i [ p(i_0 | X) , p(i_1 | X) ] = 0$$
<br>
<br>

Entretanto, atribuir os pesos {0.1, 0.1, 0.8} produziria uma previsão $ \hat{y} = 1 $:

<br>
<br>
$$ p(i_0 | X) = 0.9 x 0.1 + 0.8 x 0.1 + 0.4 x 0.8 = 0.49  $$
<br>
<br>
$$ p(i_1 | X) = 0.1 x 0.1 + 0.2 x 0.1 + 0.6 x 0.8 = 0.51  $$
<br>
<br>
$$ \hat{y} = argmax_i [ p(i_0 | X) , p(i_1 | X) ] = 1$$

# Stacking Ensemble

O *Stacking* é uma técnica de aprendizado de conjunto que combina vários modelos de classificação ou regressão por meio de um meta-classificador ou meta-regressor. Os modelos de nível básico são treinados com base em um conjunto de treinamento completo; o metamodelo é treinado nas saídas do modelo de nível básico como features.

<img src="https://blogs.sas.com/content/subconsciousmusings/files/2017/05/modelstacking.png" width="50%">
<br>

O nível básico geralmente consiste em diferentes algoritmos de aprendizado e, portanto, os conjuntos de empilhamento geralmente são heterogêneos. O algoritmo abaixo resume o empilhamento.

O *Stacking* é uma técnica comumente usada para vencer a competição de ciência de dados Kaggle. Por exemplo, o primeiro lugar no desafio da Classificação de Produtos do Grupo Otto foi conquistado por um conjunto de mais de 30 modelos cuja saída foi usada como recurso para três meta-classificadores: XGBoost, Neural Network e Adaboost. 

A ideia por trás do stacking se divide em 2 partes. Queremos treinar diversos algoritmos nos nossos dados, e queremos que as visões deles sobre os algoritmos sejam complementares, ou seja, cada algoritmo aprenda algo a partir do anterior. Desse modo, o stacking se dá por 2 loops:

 - loop externo: treino de cada algoritmo de modo independente.
 - loop interno: quebre seu dataset de treino em 10 partes, você terá 10 combinações de 9:1. Treine cada algoritmo em cada uma dessas 10 combinações. Os outputs de cada algoritmo no dataset de teste servirão de features para o próximo algoritmo no loop externo
 
De modo visual, o stacking se parece com isso:


1) Após ter feito o `train_test_split`, quebre seu dataset de treino em 10 partes


2) Fazendo combinações de 9:1 partes, treine seu algoritmo (por ex, uma Decision Tree) nas 9 partes e use a décima parte com pré-teste. Repita isso para todas as combinações. Isso ajuda o algoritmo a entender pontos de vista diferentes sob seu dataset. O nome disso é **K-Fold Cross Validation**


3) Após o treino do primeiro algoritmo em seu loop interno, faça as predictions no dataset de validação. Esse output será utilizado como *feature* para os próximos algoritmos


4) Repita o passo desse loop externo com os próximos algoritmos, até alcançar a performance desejada

O legal do Stacking é que você pode criar as features que desejar com esse método! Por exemplo, antes de fazer um regressor para prever a quantidade de vendas de um determinado produto na próxima semana, você pode antes fazer um classificador que preve se as vendas vão aumentar ou diminuir, para então passar esse output como nova feature :D

# Bagging Ensemble



*Bagging* significa *Bootstrap Aggregation*. Uma maneira de reduzir a variação de uma estimativa é fazer a média de várias estimativas. Por exemplo, podemos treinar M árvores diferentes em diferentes subconjuntos de dados (escolhidos aleatoriamente com substituição) e calcular o conjunto:

$$ f(x) = \dfrac{1}{M} ∑f_m (x)  $$

*Bagging* usa amostragem de autoinicialização para obter os subconjuntos de dados para treinar os *base learners*. Para agregar os resultados dos *base learners*, o *Bagging* usa o voto para classificação e a média para regressão.

Combinar *base learners* estáveis é menos vantajoso, pois o conjunto não ajudará a melhorar o desempenho da generalização. Uma classe comumente usada de algoritmos de conjuntos são *Random Forrest*.

A ideia por trás do *Bagging* é quebrar nosso dataset em várias partes (bags), e treinar os modelos de modo independente em cada um deles. A premissa por trás disso é que estamos utilizando modelos mais simples, sob partes dos dados, que juntos irão funcionar de modo complementar.

1) Primeiramente, nós quebramos nosso dataset em tantos bags desejarmos. O nome para isso é *bootstrapping* <br>
2) Depois, treinamos cada sub-modelo sob cada bag <br>
3) O output final é dado pela combinação dos sub-outputs. Ela pode ser feita com qualquer método de Voting

# Random Forrest


Em *Random Forrest*, cada árvore do conjunto é construída a partir de uma amostra retirada com substituição (ou seja, uma amostra de autoinicialização) do conjunto de treinamento. Além disso, em vez de usar todos os recursos, um subconjunto aleatório de recursos é selecionado, randomizando ainda mais a árvore. Como resultado, o viés da floresta aumenta um pouco, mas devido à média de árvores menos correlacionadas, sua variância diminui, resultando em um modelo geral melhor.

Em um algoritmo de árvores extremamente aleatórias, a aleatoriedade vai um passo além: os limites de divisão são aleatórios. Em vez de procurar o limite mais discriminativo, os limites são traçados aleatoriamente para cada recurso candidato e o melhor desses limites gerados aleatoriamente é escolhido como regra de divisão. Isso geralmente permite reduzir um pouco mais a variância do modelo, às custas de um pequeno aumento no viés.

A *Random Forrest* ou **Floresta Aleatória**, como o próprio nome indica, consiste em um grande número de árvores de decisão individuais que funcionam como um conjunto. Cada árvore individual na floresta aleatória cospe uma previsão de classe e a classe com mais votos se torna a previsão do nosso modelo.




## Visualização

Visualização de um modelo aleatório de floresta Fazendo uma previsão:

O conceito fundamental por trás da Floresta Aleatória é simples, mas poderoso - a sabedoria das multidões. Em termos de ciência de dados, a razão pela qual o modelo aleatório de floresta funciona tão bem é: Um grande número de modelos (árvores) relativamente não correlacionados que operam como um comitê superará qualquer um dos modelos constituintes individuais.

A baixa correlação entre os modelos é a chave. Assim como os investimentos com baixas correlações (como ações e títulos) se juntam para formar um portfólio maior que a soma de suas partes, os modelos não correlacionados podem produzir previsões de conjuntos mais precisas do que as previsões individuais. 

A razão para esse maravilhoso efeito é que as árvores se protegem de seus erros individuais (desde que nem sempre erram na mesma direção). Enquanto algumas árvores podem estar erradas, muitas outras estão certas, então, como um grupo, as árvores podem se mover na direção correta. Portanto, os pré-requisitos para um desempenho aleatório da floresta aleatória são:

- É necessário que haja algum sinal real em nossos recursos, para que os modelos criados com esses recursos sejam melhores do que a adivinhação aleatória.


- As previsões (e, portanto, os erros) feitos pelas árvores individuais precisam ter baixas correlações entre si.


### Garantir que os modelos se diversifiquem

Então, como a floresta aleatória garante que o comportamento de cada árvore individual não esteja muito correlacionado com o comportamento de qualquer outra árvore no modelo? Ele usa os dois métodos a seguir:

***Bagging (Agregação de Bootstrap)*** - As árvores de decisões são muito sensíveis aos dados nos quais são treinadas - pequenas alterações no conjunto de treinamento podem resultar em estruturas de árvore significativamente diferentes. A floresta aleatória tira vantagem disso, permitindo que cada árvore individual faça uma amostragem aleatória do conjunto de dados com substituição, resultando em árvores diferentes. Esse processo é conhecido como *Bagging*.

Observe que, com o *Bagging*, não estamos agrupando os dados de treinamento em blocos menores e treinando cada árvore em um bloco diferente. Em vez disso, se tivermos uma amostra do tamanho N, ainda estaremos alimentando cada árvore com um conjunto de treinamento do tamanho N (a menos que especificado de outra forma). Mas, em vez dos dados de treinamento originais, coletamos uma amostra aleatória do tamanho N com substituição. Por exemplo, se nossos dados de treinamento foram [1, 2, 3, 4, 5, 6], podemos fornecer a uma de nossas árvores a seguinte lista [1, 2, 2, 3, 6, 6]. Observe que ambas as listas têm o comprimento seis e que “2” e “6” são repetidos nos dados de treinamento selecionados aleatoriamente que fornecemos à nossa árvore (porque fazemos a amostragem com substituição).

***Feature Randomness (Aleatoriedade de Variáveis)*** - Em uma árvore de decisão normal, quando é hora de dividir um nó, consideramos todos os recursos possíveis e escolhemos o que produz a maior separação entre as observações no nó esquerdo e as do nó direito. Por outro lado, cada árvore em uma floresta aleatória pode escolher apenas um subconjunto aleatório de recursos. Isso força ainda mais variação entre as árvores no modelo e, finalmente, resulta em menor correlação entre as árvores e mais diversificação.

# Prática: Árvores vs Bagging vs RFs:

Vamos começar observando o comportamento desses 3 algoritmos num cenário de regressão

````python
# importando as bibliotecas
# Disable warnings
import warnings
import numpy as np
warnings.filterwarnings('ignore')
from matplotlib import pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = 10, 6
%config InlineBackend.figure_format = 'retina' 
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.ensemble import BaggingClassifier, BaggingRegressor
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split
````

````python
# criando um conjunto de dados
n_train = 150        
n_test = 1000       
noise = 0.1

def f(x):
    x = x.ravel()
    return np.exp(-x ** 2) + 1.5 * np.exp(-(x - 2) ** 2)

def generate(n_samples, noise):
    X = np.random.rand(n_samples) * 10 - 5
    X = np.sort(X).ravel()
    y = np.exp(-X ** 2) + 1.5 * np.exp(-(X - 2) ** 2)\
        + np.random.normal(0.0, noise, n_samples)
    X = X.reshape((n_samples, 1))

    return X, y

X_train, y_train = generate(n_samples=n_train, noise=noise)
X_test, y_test = generate(n_samples=n_test, noise=noise)
````

````python
plt.figure(figsize=(10, 6))
plt.plot(X_test, f(X_test), "b", label='Test')
plt.show()
````

````python


# Vamos usar o decision tree regressor
dtree = DecisionTreeRegressor().fit(X_train, y_train)
d_predict = dtree.predict(X_test)

plt.figure(figsize=(10, 6))
plt.plot(X_test, f(X_test), "b", label='Test')
plt.scatter(X_train, y_train, c="black", s=20, label='Train')
plt.plot(X_test, d_predict, "g", lw=2, label='Predict')
plt.xlim([-5, 5])
plt.legend()
plt.title("Decision tree, MSE = %.2f" 
          % np.sum((y_test - d_predict) ** 2))
plt.show()
````

Perceba que há alta variância

````python

# Bagging com decision tree regressor
bdt = BaggingRegressor(DecisionTreeRegressor()).fit(X_train, y_train)
bdt_predict = bdt.predict(X_test)

plt.figure(figsize=(10, 6))
plt.plot(X_test, f(X_test), "b")
plt.scatter(X_train, y_train, c="black", s=20)
plt.plot(X_test, bdt_predict, "y", lw=2)
plt.xlim([-5, 5])
plt.title("Bagging for decision trees, MSE = %.2f" % np.sum((y_test - bdt_predict) ** 2));
````

````python

# Random Forest (Baggin)
rf = RandomForestRegressor(n_estimators=10).fit(X_train, y_train)
rf_predict = rf.predict(X_test)

plt.figure(figsize=(10, 6))
plt.plot(X_test, f(X_test), "b")
plt.scatter(X_train, y_train, c="b", s=20)
plt.plot(X_test, rf_predict, "r", lw=2)
plt.xlim([-5, 5])
plt.title("Random forest, MSE = %.2f" % np.sum((y_test - rf_predict) ** 2));
````

Também reduz a variância.

# Hiper-parâmetros da Random Forest

O scikit possui duas implementações de Random Forest: `RandomForestClassifier` and `RandomForestRegressor`.

### Para regressão os Hiperparâmetros são os seguintes:

   - `n_estimators` — Número de árvores na floresta.(default = 10)
   - `criterion` — A função usada para medir a qualidade de um split das árvores. Na regressão poderemos usar o re “mse” para nos referirmos ao mean squared error, que é igual a redução de variância e “mae” que refere-se ao mean absolute error (default = "mse")
   - `max_features` —  O número máximo de features a serem consideradas na hora de escolher a feature para o split. Podemos especificar um percentual de features ou escolher algumas das seguintes opções:"auto" (all features), "sqrt", "log2". (default = "auto")
   -  `max_depth` — A profundidade máxima que as árvores terão na floresta (default indica que os nós serão expandidos até que todas as folhas sejam puras ou até que todas as folhas contenham menos do que o  min_samples_split samples)
   -  `min_samples_split` — O número mínimo de amostras (dados) exigidos dentro de um nó para que possa ser realizado um novo split. Pode ser especificado como o número ou como um percentual do total de amostras (default = 2)
   -  `min_samples_leaf` — Número mínimo de amostras exigidas para formar uma folha (default = 1)
   -  `min_weight_fraction_leaf` — the minimum weighted fraction of the sum total of weights (of all the input samples) required to be at a leaf node. Samples have equal weight when sample_weight is not provided (default = 0)
   - `max_leaf_nodes` — Número máximo de folhas (default = sem restrições)
   - `min_impurity_split` — Um nó será splitado se sua impureza for maior que o threshold (default = 1е-7)
   - `bootstrap` — Flag para indicar se o bootstrap samples será utilizado (default = True)
   - `oob_score` — Flag para indicar se as árvores poderão utilizar o out-of-bag samples para estimar o R^2 on em dados não vistos (default = False)
   - `n_jobs` — Número de jobs para rodar em paralelo no fit e no predict. If -1, então o número de jobs é definido como o número de cores disponíveis (default = 1)
   - `random_state` — random_state é a semente usada para criar um gerador aleatório (default = None)
   - `verbose` — Controla a visualização do processo de construção das árvores (default = 0)





### Para classificação, os Hiprparâmetros são quase idênticos. O que difere é:

   - `criterion` — Para classificação, teremos os já comentados  “gini” para indicar Gini impurity e “entropy” para usar o ganho de informação (default = "gini")
   - `class_weight` — O peso de cada classe (por default todos os pesos são 1, mas você pode criar um dicionário com pesos ou especificar como "balanced" - Usa o inverso da frequência como pesos)
   
### O que precisamos ter maior atenção quando construirmos um modelo de RF:

- `n_estimators` 
- `criterion` 
- `max_features` 
- `min_samples_leaf` 
- `max_depth`

````python

# Divide the dataset into the input and target
X, y = df[num_cols].copy(), df["churn"]

# Initialize a stratified split of our dataset for the validation process
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Initialize the classifier with the default parameters 
rfc = RandomForestClassifier(random_state=42, n_jobs=-1)

# Train it on the training set
results = cross_val_score(rfc, X, y, cv=skf)

# Evaluate the accuracy on the test set
print("CV accuracy score: {:.2f}%".format(results.mean() * 100))
````

Agora vamos explorar como cada hiper-parâmetro influencia nossos resultados.

## Número de árvores:

````python

# Create lists to save the values of accuracy on training and test sets
train_acc = []
test_acc = []
#temp_train_acc = []
#temp_test_acc = []
trees_grid = [5, 10, 15, 20, 30, 50, 75, 100]

for ntrees in trees_grid:
    rfc = RandomForestClassifier(n_estimators=ntrees, random_state=42, n_jobs=-1)
    temp_train_acc = []
    temp_test_acc = []
    for train_index, test_index in skf.split(X, y):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y[train_index], y[test_index]
        rfc.fit(X_train, y_train)
        temp_train_acc.append(rfc.score(X_train, y_train))
        temp_test_acc.append(rfc.score(X_test, y_test))
    train_acc.append(temp_train_acc)
    test_acc.append(temp_test_acc)
    
train_acc, test_acc = np.array(train_acc), np.array(test_acc)
print("Best CV accuracy is {:.2f}% with {} trees".format(np.max(test_acc.mean(axis=1))*100, 
                                                        trees_grid[np.argmax(test_acc.mean(axis=1))]))
````

````python
plt.style.use('ggplot')

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(trees_grid, train_acc.mean(axis=1), alpha=0.5, color='blue', label='train')
ax.plot(trees_grid, test_acc.mean(axis=1), alpha=0.5, color='red', label='cv')
ax.fill_between(trees_grid, test_acc.mean(axis=1) - test_acc.std(axis=1), 
                test_acc.mean(axis=1) + test_acc.std(axis=1), color='#888888', alpha=0.4)
ax.fill_between(trees_grid, test_acc.mean(axis=1) - 2*test_acc.std(axis=1), 
                test_acc.mean(axis=1) + 2*test_acc.std(axis=1), color='#888888', alpha=0.2)
ax.legend(loc='best')
ax.set_ylim([0.88,1.02])
ax.set_ylabel("Accuracy")
ax.set_xlabel("N_estimators");
````

Como podemos observar, a partir de 30 árvores, nossa performance chegou a uma saturação. Além disso, observamos que a partir de 50 árvores,o modelo apresenta 100% de acurácia, o que pode indicar uma situação de overfit. Vamos explorar agora o efeito da profundidade:

````python
# Create lists to save accuracy values on the training and test sets
train_acc = []
test_acc = []
#temp_train_acc = []
#temp_test_acc = []
max_depth_grid = [3, 5, 7, 9, 11, 13, 15, 17, 20, 22, 24]

for max_depth in max_depth_grid:
    rfc = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1, max_depth=max_depth)
    temp_train_acc = []
    temp_test_acc = []
    for train_index, test_index in skf.split(X, y):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y[train_index], y[test_index]
        rfc.fit(X_train, y_train)
        temp_train_acc.append(rfc.score(X_train, y_train))
        temp_test_acc.append(rfc.score(X_test, y_test))
    train_acc.append(temp_train_acc)
    test_acc.append(temp_test_acc)
    
train_acc, test_acc = np.asarray(train_acc), np.asarray(test_acc)
print("Best CV accuracy is {:.2f}% with {} max_depth".format(max(test_acc.mean(axis=1))*100, 
                                                        max_depth_grid[np.argmax(test_acc.mean(axis=1))]))

````

````python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(max_depth_grid, train_acc.mean(axis=1), alpha=0.5, color='blue', label='train')
ax.plot(max_depth_grid, test_acc.mean(axis=1), alpha=0.5, color='red', label='cv')
ax.fill_between(max_depth_grid, train_acc.mean(axis=1) - train_acc.std(axis=1), 
                train_acc.mean(axis=1) + train_acc.std(axis=1), color='#888888', alpha=0.4)
ax.fill_between(max_depth_grid, train_acc.mean(axis=1) - 2*train_acc.std(axis=1), 
                train_acc.mean(axis=1) + 2*train_acc.std(axis=1), color='#888888', alpha=0.2)
ax.fill_between(max_depth_grid, test_acc.mean(axis=1) - test_acc.std(axis=1), 
                test_acc.mean(axis=1) + test_acc.std(axis=1), color='#888888', alpha=0.4)
ax.fill_between(max_depth_grid, test_acc.mean(axis=1) - 2*test_acc.std(axis=1), 
                test_acc.mean(axis=1) + 2*test_acc.std(axis=1), color='#888888', alpha=0.2)
ax.legend(loc='best')
ax.set_ylim([0.88,1.02])
ax.set_ylabel("Accuracy")
ax.set_xlabel("Max_depth");
````

Podemos observar que a profundidade da árvore possui um comportamento mais controlado em relação ao overfit. Vamos explorar min_samples_leaf:

````python
# Create lists to save accuracy values on the training and test sets
train_acc = []
test_acc = []
#temp_train_acc = []
#temp_test_acc = []
min_samples_leaf_grid = [1, 3, 5, 7, 9, 11, 13, 15, 17, 20, 22, 24]

for min_samples_leaf in min_samples_leaf_grid:
    rfc = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1, 
                                 min_samples_leaf=min_samples_leaf)
    temp_train_acc = []
    temp_test_acc = []
    for train_index, test_index in skf.split(X, y):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y[train_index], y[test_index]
        rfc.fit(X_train, y_train)
        temp_train_acc.append(rfc.score(X_train, y_train))
        temp_test_acc.append(rfc.score(X_test, y_test))
    train_acc.append(temp_train_acc)
    test_acc.append(temp_test_acc)
    
train_acc, test_acc = np.asarray(train_acc), np.asarray(test_acc)
print("Best CV accuracy is {:.2f}% with {} min_samples_leaf".format(max(test_acc.mean(axis=1))*100, 
                                                        min_samples_leaf_grid[np.argmax(test_acc.mean(axis=1))]))
````

````python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(min_samples_leaf_grid, train_acc.mean(axis=1), alpha=0.5, color='blue', label='train')
ax.plot(min_samples_leaf_grid, test_acc.mean(axis=1), alpha=0.5, color='red', label='cv')
ax.fill_between(min_samples_leaf_grid, test_acc.mean(axis=1) - test_acc.std(axis=1), 
                test_acc.mean(axis=1) + test_acc.std(axis=1), color='#888888', alpha=0.4)
ax.fill_between(min_samples_leaf_grid, test_acc.mean(axis=1) - 2*test_acc.std(axis=1), 
                test_acc.mean(axis=1) + 2*test_acc.std(axis=1), color='#888888', alpha=0.2)
ax.legend(loc='best')
ax.set_ylim([0.88,1.02])
ax.set_ylabel("Accuracy")
ax.set_xlabel("Min_samples_leaf");
````

Nesse caso, não vamos um ganho substancial na acurácia no dataset de validação, mas conseguimos nos proteger do overfitting enquanto mantemos a acurácia a ~92%.

# Pros & Cons de Random Forests

## Pros:

- Alta precisão de previsão; terá melhor desempenho do que algoritmos lineares na maioria dos problemas; a precisão é comparável à do boosting.
- Robusto aos valores extremos, graças à amostragem aleatória.
- Insensível ao dimensionamento ou escala de recursos, bem como a quaisquer outras transformações monotônicas devido à seleção aleatória do subespaço.
- Não requer ajuste fino de parâmetros, funciona muito bem fora da caixa. Com o ajuste, é possível obter um ganho de precisão de 0,5 a 3%, dependendo da configuração e dos dados do problema.
- Eficiente para conjuntos de dados com um grande número de recursos e classes.
- Lida com variáveis ​​contínuas e discretas igualmente bem.
- Raramente sobre-ajusta ou causa Overfit. Na prática, um aumento no número da árvore quase sempre melhora a composição. Mas, depois de atingir um certo número de árvores, a curva de aprendizado está muito próxima da assíntota.
- Existem métodos desenvolvidos para estimar a importância dos recursos ou Feature Importance.
- Funciona bem com dados faltantes e mantém boa precisão, mesmo quando uma grande parte dos dados está ausente.
- Fornece meios para ponderar classes em todo o conjunto de dados, bem como para cada amostra de árvore.
- Pode ser utilizada para calcular as proximidades entre pares de exemplos que podem ser usados ​​subsequentemente em cluster, detecção de outlier ou representações de dados interessantes.
- A funcionalidade e as propriedades acima podem ser estendidas para dados não rotulados, para permitir cluster não supervisionado, visualização de dados e detecção de outlier.
- Facilmente paralelizado e altamente escalável.
 


## Cons:

- Em comparação com uma única árvore de decisão, a saída da Random Forest é mais difícil de interpretar.
- Não há valores p formais para a estimativa da significância das características.
- Executa métodos piores que os lineares no caso de dados esparsos: entradas de texto, conjunto de palavras, etc.
- Diferentemente da regressão linear, a Random Forest não pode extrapolar. Porém, isso também pode ser considerado uma vantagem, porque os Outliers não causam valores extremos nas florestas aleatórias.
- Propenso a Overfit em alguns problemas, especialmente quando se lida com dados ruidosos.
- No caso de variáveis categóricas com números de níveis variados, florestas aleatórias favorecem variáveis com um número maior de níveis. A árvore se ajustará mais a um recurso com muitos níveis, porque isso ganha maior precisão.
- Se um conjunto de dados contiver grupos de recursos correlatos com importância semelhante para as classes previstas, a preferência será dada a grupos menores.
- O modelo resultante é grande e requer muita RAM.
