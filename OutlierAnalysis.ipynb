# Introdução

> “Observation which deviates so much from other observations as to arouse suspicion it was generated by a different mechanism” — Hawkins(1980)

Quando falamos sobre outliers, nos referimos a valores (dados, linhas de dataframe) extremos que se desviam de outras observações nos dados. O surgimento desses outliers podem indicar diversos fenômenos, desde a variabilidade das medições realizadas, erros experimentais, ou até mesmo uma nova descoberta.

<img src="https://miro.medium.com/max/732/1*J-ds9KYQBheiBaIJn78seg.gif">

**Tipos de Outlier**

Outliers podem ser de dois tipos: <br>
**Univariados**: Outliers univariados podem ser encontrados ao observar uma distribuição de valores em uma única feature.

**Multivariados**: Outliers multivariados podem ser encontrados em um espaço n-dimensional (de n-features). Olhar para distribuições em espaços n-dimensionais pode ser muito difícil para o cérebro humano, por isso precisamos treinar um modelo para fazer isso por nós.

Além disso, Outliers também podem ser classificados de formas diferentes, como: Pontual Outliers, Contextual Outliers ou Collective Outliers. 

- Pontual Outliers: São pontos de dados únicos que ficam longe do resto da distribuição. São também conhecidos como outliers globais. <br>
<img src="https://www.snl.com/articles/404226244.png">
- Contextual Outliers: Os valores discrepantes contextuais podem ser ruído nos dados. São pontos de dados que, se analisados em um outro contexto, poderiam ser "normais".
<img src="https://2b309c36t6402pmjhp3206t0-wpengine.netdna-ssl.com/wp-content/uploads/outliers4-2.png"><br>
- Collective Outliers: Quando um subconjunto de pontos possuem comportamento semelhante, mas que difere de todos os demais. 
<img src="https://www.programmersought.com/images/330/50858dbaa8366fec5d5b1a9b0515c382.png">

Para tangibilizar mais, imagine que um avião precisou pousar numa rodovia. Esse seria um exemplo de Pontual Outlier, porque é um evento verdadeiramente raro um avião ter que pousar lá. Agora, imagine a marginal pinheiros. Para quem mora em SP deve tê-la imaginado em um congestionamento, como de costume. Porém, se isso estivesse acontecendo às 3 da manhã, quando o tráfego geralmente não começa até o final da manhã, quando as pessoas estão indo para o trabalho, isso poderia ser considerado um Contextual Outlier. Por fim, se todos os carros na rodovia estivessem se movendo para a pista da esquerda ao mesmo tempo, isso seria um Collective Outlier, porque embora definitivamente não seja raro que as pessoas se movam para a pista da esquerda, é incomum que todos os carros mudem de lugar exatamente ao mesmo tempo .

# Origens

- Erros de inserção:
   É fruto do erro humano. Por exemplo, uma companhia de água que realiza medições em hidrômetros de casa em casa. 
   Em um dado momento, o operador pode anotar errado o valor do consumo daquela casa e reportar um valor muito discrepante das demais observações históricas de consumo.
   
- Erros de Medição:
    O erro de inserção é mais sistemático, pois tende a ocorrer de forma conjunta para mais observações. Por exemplo,imagine um medidor de velocidade que está descalibrado. Ele pode estar fornecendo velocidades com um viés de 20km/h, por falta de calibração. Ou seja, todas as velocidades medidas terão esse acréscimo inicial de 20km/h. Algumas dessa observações se tornarão outliers por isso. 
    
- Erros experimentias:
    São erros que ocorrem durante a execução/planejamento de um experimento.Um exemplo é o de uma reação química que produz elementos que impedem ou introduzem ruído na hora da leitura dos sensores.


- Erros de processamento: São outliers introduzidos durante o processo de processamento dos dados. É resultado de alguma operação, cálculo e etc.

- Descobertas: São dados que desviam do padrão, mas porque representam fenômenos novos ou novas condições de processo e etc.

No processo de produção, coleta, processamento e análise de dados, os valores outliers podem vir de muitas fontes e se esconder em muitas dimensões. Aquelas que não são produto de um erro são geralmente novos fenômenos ou comportamentos.
A detecção de outliers é de grande importância para quase todas as disciplinas quantitativas (ou seja: Física, Economia, Finanças, Aprendizado de Máquina, Segurança Cibernética). No aprendizado de máquina e em qualquer disciplina quantitativa, a qualidade dos dados é tão importante quanto a qualidade de um modelo de predição ou classificação.

Os algoritmos de aprendizado de máquina são sensíveis ao intervalo e à distribuição dos valores dos atributos. Dados discrepantes podem prejudicar e enganar o processo de treinamento, resultando em tempos de treinamento mais longos, modelos menos precisos e, em última análise, resultados mais fracos.

Quando ocorre algum problema no modelo devido a esses outliers, os chamamos de "influential points".
Mas professor, quer dizer então que às vezes eles não vão influenciar em nada? SIM!

Um dado pode ser outlier, mas pode não ser influential.

Esse pequeno fato traz pro jogo uma das perguntas mais feitas pelos estudantes de ciência de dados: **PRECISO DROPAR OS OUTLIERS?**


# Dropar ou não dropar, eis a questão

Valores discrepantes são um daqueles problemas estatísticos que todos conhecem, mas a maioria das pessoas não tem certeza de como lidar com eles. A maioria das estatísticas paramétricas, como médias, desvios padrão e correlações, e todas as estatísticas com base nelas, são altamente sensíveis a outliers. E como as suposições de procedimentos estatísticos comuns, como regressão linear e ANOVA, também são baseadas nessas estatísticas, os valores discrepantes podem realmente bagunçar sua análise.

Apesar de tudo isso, por mais que você queira, NÃO é aceitável abandonar uma observação apenas porque é um outlier. Elas podem ser observações legítimas e, às vezes, as mais interessantes. É importante investigar a natureza do outlier antes de decidir.

Note que as fraudes são dados à parte dos demais. Ou seja, se por meio de detecção de outlier, podemos melhor a performance de um algoritmo de fraude. Nesse caso, não faz sentido dropar os outliers, pois eles são o objeto de estudo.

A detecção de outliers também é usada em análise de texto. Por exemplo, uma plataforma de eventos pode utilizar a análise de outliers para identificar novos temas de evento surgindo na plataforma.

- Se for óbvio que o outlier é devido a dados incorretamente inseridos ou medidos, você deve descartar o outlier:
Por exemplo, ao analisar um conjunto de dados em que o peso de uma mulher era registrado como 19 kg. Seu verdadeiro peso era provavelmente 91 até mesmo 119, mas como eu não sabemos qual, deixamos lado o outlier. 
Isso também se aplica a uma situação na qual você sabe que o dado não mediu com precisão o que você pretendia. Por exemplo, se você está testando o tempo de reação das pessoas a um evento, mas viu que o participante não está prestando atenção e apertando aleatoriamente a tecla de resposta, você sabe que não é uma medida precisa.

- Se o valor discrepante não alterar os resultados, mas afetar as suposições, você pode dropar.

- Mais comumente, os outtliers afetarão as hipóteses e também os resultados, métricas e etc. Nessa situação, é interessante reportar ambos os resultados: com e sem os outliers.

# Outlier x High Leverage x Influential Points

Ao longo do curso e em diversas situações utilizamos uma única denominação para dados discrepantes, que não possuam o comportamento esperado: **Outliers**

Contudo, há distinção entre os dados considerados "anormais". 

- Um dado é considerado outlier, quando é discrepante em Y. Ou seja, quando nosso target possui um comportamento anômalo em algum ponto de seu domínio, aquele ponto é chamado de outlier.
<img src="https://blogdotrichanchordotcom.files.wordpress.com/2016/05/outliers.png?w=640">

O caso acima nos traz essa situação. Obviamente, isso pode ser generalizado para mais dimensões. Imagine que o eixo vertical é o eixo Y  e nosso target.

- Além de outlier, um dado pode ser considerado High Leverage.

Na imagem abaixo, em termos de Y, nosso target, temos uma discrepância, mas o ponto ainda segue o comportamento dos demais. Não podemos considerá-lo um outlier. Contudo, note que em relação a X, também temos um valor discrepante. Quando isso ocorre, podemos denominar esse ponto de high leverage. É importante notar que, para esse caso, nosso modelo de regressão linear não seria afetado por esse high leverage point.

<img src="https://online.stat.psu.edu/stat462/sites/onlinecourses.science.psu.edu.stat462/files/09influential/influ_scatter_06/index.png">

- Por fim, temos o chamado Influential point, que será capaz de alterar nosso modelo, influenciar em sua performance e etc.

<img src="https://learnche.org/pid/_images/influence-of-outliers.png">

Note os modelos acima. Temos o Model A, com um outlier à esquerda da reta de regressão. Esse influential point traz para perto de si a reta, aumentando o coeficiente angular.  O Model B possui um influential point à direita da reta. Note que esse influential point traz para perto de si a reta e, dessa forma, tende a diminuir o coeficiente angular. Por fim, o modelo sem os influential points não sofre com o efeito de "gravidade".

# Isolation Forests

Existem vários algoritmos de detecção de outlier. Já vimos DBSCAN, por exemplo, que atribui -1 aos pontos de dados que não estão em regiões densas, ou seja, não são core points nem border points. Outros algoritmos bastante famosos são o one-class svm e o Local Outlier Factor. Hoje veremos o funcionamento e implementação de um outro algoritmo, chamado de Isolation Forest.

O Isolation Forest é um algoritmo simples, com poucos hiperparâmetros. Aliás, esses hiperparâmetros podem ser otimizados por validação cruzada também e já sabemos fazer isso.

Como o próprio nome sugere, é um algoritmo baseado em árvores de decisão. Por se tratar de um conjunto de modelos, várias ávores, também é conhecido como ensemble. Vamos estudar ensembles mais profundamente na próxima aula.

Nessas árvores, as partições são criadas primeiro selecionando aleatoriamente uma feature e, em seguida, selecionando um valor de divisão aleatório entre o valor mínimo e máximo do recurso selecionado.

Até que se prove o contrário, outliers são menos frequentes do que observações regulares e são diferentes deles em termos de valores (eles estão mais distantes das observações regulares no espaço). É por isso que, usando esse particionamento aleatório, eles devem ser identificados mais perto da raiz da árvore (menor comprimento médio do caminho, ou seja, o número de arestas que uma observação deve passar na árvore indo da raiz ao nó terminal), com menos divisões necessárias.

Falei grego?

<img src="https://opendistro.github.io/for-elasticsearch/assets/media/blog-images/isolationforest.png">

Para entender melhor faço a seguinte pergunta: Qual ponto tem maior probabilidade de ser "isolado" dos demais, mais rapidamente (com menos cortes), com cortes aleatórios? Isso mesmo, os pontos mais discrepantes, longe da multidão de dados, tem uma área não prenchida muito grande ao redor. Em qualquer região desse espaço em branco que cortemos os nossos dados, já conseguiremos isolar nosso ponto vermelho. 

<img src="https://miro.medium.com/max/1400/1*EBnpXIwECYXW1BcJ0mSxqw.png">

O isolation forest se baseia nisso! Ele calcula quantos cortes em média é necessário para se isolar um ponto específico. De forma geral, os pontos discrepantes precisam de menos cortes. Agora ele pode usar essa informação para detectar os outliers. O Isolation Forest utiliza várias árvores e utiliza a média de número de cortes para gerar um score de outlier ou mesmo uma decisão 1 ou -1.

<img src="https://miro.medium.com/max/1400/1*_und72DAAsXQs3E70sYOjg.gif">

Criando um Isolation Forest

Primeiro, iremos criar um dataset de exemplo e também criaremos outliers.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import savefig
from sklearn.ensemble import IsolationForest

# default plot settings
plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.figsize'] = [15, 10]
# Generating data ----

rng = np.random.RandomState(42)

# Generating training data 
X_train = 0.2 * rng.randn(1000, 2)
X_train = np.r_[X_train + 3, X_train]
X_train = pd.DataFrame(X_train, columns = ['x1', 'x2'])

# Generating new, 'normal' observation
X_test = 0.2 * rng.randn(200, 2)
X_test = np.r_[X_test + 3, X_test]
X_test = pd.DataFrame(X_test, columns = ['x1', 'x2'])

# Generating outliers
X_outliers = rng.uniform(low=-1, high=5, size=(50, 2))
X_outliers = pd.DataFrame(X_outliers, columns = ['x1', 'x2'])
```

```python
X_outliers.shape
```

Temos 50 outliers para testar o algoritmo.

```python
# Plotting generated data ----

plt.title("Data")

p1 = plt.scatter(X_train.x1, X_train.x2, c='white',
                 s=20*4, edgecolor='k')
p2 = plt.scatter(X_test.x1, X_test.x2, c='green',
                 s=20*4, edgecolor='k')
p3 = plt.scatter(X_outliers.x1, X_outliers.x2, c='red',
                s=20*4, edgecolor='k')

plt.axis('tight')
plt.xlim((-2, 5))
plt.ylim((-2, 5))
plt.legend([p1, p2, p3],
           ["training observations",
            "new regular obs.", "new abnormal obs."],
           loc="lower right")
plt.show()
```

Para o Isolation Forest, um parâmetro interessante é o contamination, que indica o percentual de outliers no dataset.

```python
# Isolation Forest ----

# training the model
clf = IsolationForest(contamination=0.05, random_state=rng)
clf.fit(X_train)

# predictions
y_pred_train = clf.predict(X_train)
y_pred_test = clf.predict(X_test)
y_pred_outliers = clf.predict(X_outliers)
```

```python

# new, 'normal' observations
print("Accuracy:", list(y_pred_train).count(1)/y_pred_train.shape[0])
```

```python

# new, 'normal' observations
print("Accuracy:", list(y_pred_test).count(1)/y_pred_test.shape[0])
```

```python
# outliers
print("Accuracy:", list(y_pred_outliers).count(-1)/y_pred_outliers.shape[0])
```
