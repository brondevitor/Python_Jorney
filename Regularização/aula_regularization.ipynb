# Regulariza√ß√£o em Machine Learning

Um dos principais aspectos do treinamento do seu modelo de aprendizado de m√°quina √© evitar o overfitting, pois neste caso o modelo ter√° uma baixa precis√£o. Isso acontece porque o seu modelo dificilmente ir√° conseguir capturar o ru√≠do em seu conjunto de dados de treinamento. Por ru√≠do, queremos dizer os pontos de dados que realmente n√£o representam as propriedades reais de seus dados, mas de chance aleat√≥ria. Aprender esses pontos de dados torna o seu modelo mais flex√≠vel, sob o risco de overfitting.

O conceito de balanceamento de vi√©s e vari√¢ncia √© √∫til para entender o fen√¥meno do overfitting e varia√ß√£o de balanceamento, para controlar erros do Machine Leanrning.

## Formas de Regulariza√ß√£o

Vamos abordar aqui duas formas de se fazer a regulariza√ß√£o:

I. Aumentar o conjunto de dados (n√∫mero de observa√ß√µes) \
II. Penaliza√ß√£o dos otimizadores

## Aumento do conjunto de dados

O Overfitting pode ser controlado aumentando o tamanho do conjunto de dados de treinamento. Aumentar o tamanho do conjunto de dados para evitar ajustes excessivos se refere ao aumento do n√∫mero de observa√ß√µes (ou linhas) e n√£o do n√∫mero de recursos (ou colunas). A adi√ß√£o de colunas pode levar ao aumento da complexidade do problema e, portanto, pode resultar em um pior desempenho. A seguir veremos um exemplo de como aumentar um conjunto de dados para Regularizar um modelo

## Penaliza√ß√£o de otimizadores

Regulariza√ß√£o √© uma t√©cnica que restringe/regulariza/reduz as estimativas dos coeficientes, diminuiindo seus valores, que ficam portanto mais pr√≥ximos de zero. Em outras palavras, esta t√©cnica desencoraja a aprendizagem de um modelo mais complexo ou flex√≠vel, de modo a evitar o risco de overfitting.

Uma rela√ß√£o simples de regress√£o linear se parece com a f√≥rmula abaixo, onde $y$ representa a rela√ß√£o aprendida e ùúÉ representa as estimativas de coeficiente para diferentes vari√°veis ou preditores (x).


<br>
<br>
$$y = \theta_0 + \theta_1x_1 + \theta_2x_2 + \theta_3x_3 + ... + \theta_nx_n$$
<br>
<br>

O procedimento de ajuste envolve uma fun√ß√£o de perda, conhecida como soma residual de quadrados ou RSS, sendo que os coeficientes s√£o escolhidos, de forma que minimizem essa fun√ß√£o de perda.

A regulariza√ß√£o ajustar√° os coeficientes com base nos seus dados de treinamento. Se houver ru√≠do nos dados de treinamento, os coeficientes estimados n√£o ser√£o generalizados nos dados futuros. √â aqui que a regulariza√ß√£o entra e encolhe ou regulariza essas estimativas aprendidas para zero.

## 1. Ridge Regression

<br>
<br>
$$Ridge : \lambda \sum_{i=1}^{n} \theta_i^2$$
<br>
<br>

A f√≥rmula acima mostra o coeficiente de Ridge que modifica a regress√£o pela adi√ß√£o da quantidade de contra√ß√£o. Agora, os coeficientes s√£o estimados minimizando essa fun√ß√£o, e o Œª √© o par√¢metro de sintonia que decide quanto queremos penalizar a flexibilidade do nosso modelo. O aumento na flexibilidade de um modelo √© representado pelo aumento de seus coeficientes e, se quisermos minimizar a fun√ß√£o acima, esses coeficientes precisam ser pequenos. √â assim que a t√©cnica de regress√£o de Ridge impede que os coeficientes subam demais. Al√©m disso, observe que encolhemos a associa√ß√£o estimada de cada vari√°vel com a resposta, exceto o intercepto $\theta_0$. 

Quando Œª = 0, o termo de penalidade n√£o tem efeito, e as estimativas produzidas por regress√£o de rebordo ser√£o iguais a m√≠nimos quadrados. No entanto, como Œª ‚Üí ‚àû, o impacto da penalidade de contra√ß√£o aumenta, e as estimativas coeficientes da regress√£o da crista se aproximam de zero. Como pode ser visto, selecionar um bom valor de Œª √© cr√≠tico. A valida√ß√£o cruzada √© √∫til para esse prop√≥sito. As estimativas dos coeficientes produzidas por este m√©todo s√£o tamb√©m conhecidas como norma L2.

Os coeficientes que s√£o produzidos pelo m√©todo dos m√≠nimos quadrados padr√£o s√£o escala equivariante, ou seja, se multiplicarmos cada entrada por c, ent√£o os coeficientes correspondentes s√£o escalonados por um fator de 1 / c. Portanto, independentemente de como o preditor √© escalado, a multiplica√ß√£o de preditor e coeficiente permanece a mesma. No entanto, esse n√£o √© o caso da regress√£o de Ridge e, portanto, precisamos padronizar os preditores ou trazer os preditores para a mesma escala antes de executar a regress√£o de Ridge.


## 2. Lasso Regression

<br>
<br>
$$Lasso : \lambda \sum_{i=1}^{n} |\theta_i| $$
<br>
<br>

Lasso √© outra varia√ß√£o de regulariza√ß√£o, em que a fun√ß√£o acima √© minimizada. √â claro que essa varia√ß√£o difere da regress√£o de Ridge apenas em penalizar os altos coeficientes. Ele usa | $\theta_i$ | (m√≥dulo) em vez de quadrados de $\theta_i$, como sua penalidade. Nas estat√≠sticas, isso √© conhecido como a norma L1.

Vamos dar uma olhada nos m√©todos acima com uma perspectiva diferente: a regress√£o de Ridge pode ser pensada como resolvendo uma equa√ß√£o, onde a soma dos quadrados dos coeficientes √© menor ou igual a s. E o Lasso pode ser pensado como uma equa√ß√£o onde a soma do m√≥dulo de coeficientes √© menor ou igual a s. Aqui, s √© uma constante que existe para cada valor do fator de encolhimento Œª. Essas equa√ß√µes tamb√©m s√£o chamadas de fun√ß√µes de restri√ß√£o.

Considere os seus dois par√¢metros em um determinado problema. Ent√£o, de acordo com a formula√ß√£o acima, a regress√£o de Ridge √© expressa por $\theta_1^2 +  \theta_2^2 ‚â§ s$. Isto implica que os coeficientes de regress√£o de Ridge t√™m a menor RSS (fun√ß√£o de perda) para todos os pontos que se encontram dentro da circunfer√™ncia dada por $\theta_1^2 +  \theta_2^2 ‚â§ s$.

Da mesma forma, para Lasso, a equa√ß√£o se torna $ |\theta_1| +  |\theta_2| ‚â§ s $. Isto implica que os coeficientes de Lasso possuem o menor RSS (fun√ß√£o de perda) para todos os pontos que est√£o dentro do diamante dado por $ |\theta_1| +  |\theta_2| ‚â§ s $.

A imagem acima mostra as fun√ß√µes de restri√ß√£o (√°reas verdes), para Lasso (esquerda) e regress√£o de Ridge (direita), juntamente com contornos para RSS (elipse vermelha). Pontos na elipse compartilham o valor do RSS. Para um valor muito grande de s, as regi√µes verdes conter√£o o centro da elipse, fazendo estimativas de coeficiente de ambas as t√©cnicas de regress√£o, iguais √†s estimativas de m√≠nimos quadrados. Mas, este n√£o √© o caso na imagem acima. Nesse caso, as estimativas do coeficiente de regress√£o de Lasso e Ridge s√£o fornecidas pelo primeiro ponto em que uma elipse entra em contato com a regi√£o de restri√ß√£o. Como a regress√£o de Ridge possui uma restri√ß√£o circular sem pontos agudos, essa interse√ß√£o geralmente n√£o ocorre em um eixo e, portanto, as estimativas do coeficiente de regress√£o de Ridge ser√£o exclusivamente diferentes de zero. No entanto, a restri√ß√£o de Lasso tem cantos em cada um dos eixos e, portanto, a elipse geralmente cruza a regi√£o de restri√ß√£o em um eixo. Quando isso ocorre, um dos coeficientes ser√° igual a zero. Em dimens√µes mais altas (onde os par√¢metros s√£o muito maiores que 2), muitas das estimativas de coeficiente podem ser iguais a zero simultaneamente.

Isso esclarece a desvantagem √≥bvia da regress√£o de Ridge, que √© a interpretabilidade do modelo. Reduzir√° os coeficientes dos preditores menos importantes, muito pr√≥ximos de zero. Mas isso nunca os tornar√° exatamente zero. Em outras palavras, o modelo final incluir√° todos os preditores. No entanto, no caso do Lasso, a penalidade de L1 tem o efeito de for√ßar algumas das estimativas de coeficiente a serem exatamente iguais a zero quando o par√¢metro de ajuste Œª √© suficientemente grande. Portanto, o m√©todo do Lasso tamb√©m realiza sele√ß√£o de vari√°veis ‚Äã‚Äãe √© dito que produz modelos escassos.

## 3. ElasticNet Regression (L1 + L2)

## O que a Regulariza√ß√£o faz exatamente?

Um modelo padr√£o de m√≠nimos quadrados tende a apresentar alguma varia√ß√£o, ou seja, esse modelo n√£o generaliza bem para um conjunto de dados diferente dos dados de treinamento. **A regulariza√ß√£o reduz significativamente a vari√¢ncia do modelo, sem aumento substancial de seu vi√©s.** Portanto, o par√¢metro de ajuste Œª, usado nas t√©cnicas de regulariza√ß√£o descritas acima, controla o impacto na polariza√ß√£o e varia√ß√£o. √Ä medida que o valor de Œª aumenta, reduz o valor dos coeficientes e, assim, reduz a varia√ß√£o. At√© certo ponto, esse aumento em Œª √© ben√©fico, pois reduz apenas a varia√ß√£o (evitando, portanto, o ajuste excessivo), sem perder propriedades importantes nos dados. Mas, ap√≥s um certo valor, o modelo come√ßa a perder propriedades importantes, dando origem a um vi√©s no modelo e, portanto, a um ajuste insuficiente. Portanto, o valor de Œª deve ser cuidadosamente selecionado.

# Regress√£o Ridge Te√≥rica

Apresentaremos agora uma classe que realiza a Regulariza√ß√£o de Ridge e faremos a aplica√ß√£o deta regulariza√ß√£o em dados sint√©ticos.

## Classe de Regulariza√ß√£o Ridge

Nesta classe de modelo de regress√£o Ridge, podemos identificar o elemento `I` da Regulariza√ß√£o, que √© baseado no hiper-par√¢metro`l` sendo adicionado na modelagem para regularizar o c√°lculo dos pesos ou estimadores `w_hat`, que s√£o utilizados na predi√ß√£o.

````python
class ridge_regression(object):

    def __init__(self, l=1e-4):
        self.l = l

    def fit(self, X_train, y_train):
        # adiciona coluna de 1 nos dados
        X = np.insert(X_train, 0, 1, 1)

        # estima os w_hat
        I = np.identity(X.shape[1]) * self.l # multiplicando o coeficiente de regulariza√ß√£o l pela Identidade
        w_hat = np.dot( np.dot( np.linalg.inv(np.dot(X.T, X) + I), X.T), y_train) # termo de regulariza√ß√£o: I
        
        # calcula os estimadores
        self.w_hat = w_hat
        self.coef = self.w_hat[1:]
        self.intercept = self.w_hat[0]

    def predict(self, X_test):
        X = np.insert(X_test, 0, 1, 1) # adiciona coluna de 1 nos dados
        y_pred = np.dot(X, self.w_hat) # X * w_hat = y_hat
        return y_pred
````

## Aplica√ß√£o

Regulariza√ß√£o de uma fun√ß√£o de ajuste √† dados sit√©ticos gerados a prtir de uam fun√ß√£o quadr√°tica, onde o hiper-par√¢metro de regulariza√ß√£o √© aumentado, generalizando a fun√ß√£o de otimiza√ß√£o de uma regress√£o com overfit para uma regress√£o gen√©rica. o Plot do gr√°fico das fuu√ß√µes de otimiza√ß√£o com o aumento do hiper-par√¢metro `l`demonstram como a otimiza√ß√£o gen√©rica fica liver da vari√¢ncia dos dados de treino.

````python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
plt.figure(figsize=(20,6))
# definindo uma fun√ß√£o quadr√°tica para gera√ß√£o dos dados sint√©ticos
def f(x):
    return x**2 - 10*x + 5 + np.random.normal(0,10,size=x.shape)

# Gera os dados sint√©ticos com a fun√ß√£o quadr√°tica
x_train = np.random.uniform(0,15,15)
y = f(x_train)
X = x_train[:, np.newaxis]

# par√¢metros do gr√°fico
colors = ['blue', 'green', 'orange']
lw = 2

# ajusta polin√¥mios de grau 10 com diferentes coeficinetes de regulariza√ß√£o
for count, l in enumerate([1e-8, 1e-3, 1e2]):
    
    # crias as vari√°veis polinomiais e as passa pelo modelo de regress√£o de ridge
    model = make_pipeline(PolynomialFeatures(10), ridge_regression(l=l))
    model.fit(x_train.reshape(-1,1), y)
    
    # coloca as previs√µes no gr√°fico
    x_plot = np.linspace(0,15, 50)
    y_plot = model.predict(x_plot.reshape(-1,1))
    plt.plot(x_plot.reshape(-1,1), y_plot, color=colors[count], linewidth=lw,
             label="$\lambda =$ %.1e" % l)

plt.scatter(x_train, y, color='navy', s=30, marker='o')
plt.ylim([-60,90])
plt.legend(loc='best')
plt.show()
````

## Regulariza√ß√£o 

### Sample dos dados

Inicilamente vamos diminuir a quantidade de dados por cinco, rogar uma regress√£o linear e medir os resultados de treino e teste (RMSE de treino e teste). Ent√£o calculamos a diferen√ßa para verificar a perda pelo sobrajustge dos dados. 

Depois disso, vamos trabalhar com o DataSet com o tamanho original, o que para n√≥s representa uma regulariza√ß√£o, e repetir o processo para comparar a diferen√ßa entre as diferen√ßas de metrica entre treino e teste dos dados de Sample e dos dados completos ou Regularizados.

Esperamos que a diferen√ßa entre as m√©tricas RMSE de treino e teste seja menos para os dados Regularizados, o que significa que temos um modelo mais gen√©rico, ou seja, menos viciado nos dados de treino (ou com menos vari√¢ncia), e capaz de entender melhor os padr√µes dos dados de teste ou dados desconhecidos.

## Diminuindo o tamanho dos dados em n vezes

````python
# retirando uma amostra do DataSet
n=3
df_sample = df.sample(n=int(df.shape[0]/n), random_state=0)
df_sample.shape
````

### Split dos dados SAMPLE em treino e teste

````python
from sklearn.model_selection import train_test_split
# split dos dados
train, test = train_test_split(df_sample, test_size = 0.2, random_state=0)
````

### Ajuste do modelo SAMPLE por uma regress√£o linear

````python
# criando uma fun√ß√£o com as vari√°veis e rodando uma regress√£o linear
function = 'PRICE ~ ZN + CHAS + RM + RAD + \
                    CRIM + AGE + PTRATIO + B + LSTAT + \
                    INDUS + NOX + DIS + TAX'
# instanciando, ajustando e imprimindo os resultados do modelo OLS
model = smf.ols(function, train).fit()
print(model.summary())
````

### Medindo os resultados SAMPLE com a diferen√ßa entre treino e teste

````python
# importando as bibliotecas
from sklearn.metrics import mean_squared_error
from math import sqrt

# fazendo predi√ß√µes de treino
y_hat_train = model.predict(train)
# calculando a raiz quadrada do erro m√©dio quadr√°tico
rmse_train = sqrt(mean_squared_error(train.PRICE, y_hat_train))
print('RMSE de Treino:', rmse_train)

# fazendo predi√ß√µes de teste
y_hat_test = model.predict(test)
# calculando a raiz quadrada do erro m√©dio quadr√°tico
rmse_test = sqrt(mean_squared_error(test.PRICE, y_hat_test))
print('RMSE de Teste:', rmse_test)

print('Diferen√ßa:', rmse_test-rmse_train)
````

### Split dos dados REGULARIZADOS em treino e teste

````python
# split dos dados
train, test = train_test_split(df, test_size = 0.2, random_state=99)
````

### Ajuste do modelo REGULARIZADO por uma regress√£o linear

````python
# criando uma fun√ß√£o com as vari√°veis e rodando uma regress√£o linear
function = 'PRICE ~ ZN + CHAS + RM + RAD + \
                    CRIM + AGE + PTRATIO + B + LSTAT + \
                    INDUS + NOX + DIS + TAX'
# instanciando, ajustando e imprimindo os resultados do modelo OLS
model = smf.ols(function, train).fit()
print(model.summary())
````

### Medindo os resultados REGULARIZADOS com a diferen√ßa entre treino e teste

````python
# fazendo predi√ß√µes de treino
y_hat_train = model.predict(train)
# calculando a raiz quadrada do erro m√©dio quadr√°tico
rmse_train = sqrt(mean_squared_error(train.PRICE, y_hat_train))
print('RMSE de Treino:', rmse_train)

# fazendo predi√ß√µes de teste
y_hat_test = model.predict(test)
# calculando a raiz quadrada do erro m√©dio quadr√°tico
rmse_test = sqrt(mean_squared_error(test.PRICE, y_hat_test))
print('RMSE de Teste:', rmse_test)

print('Diferen√ßa:', rmse_test-rmse_train)
````

## Conclus√£o 

Como esperado, a diferen√ßa entre as m√©tricas de treino e teste para o modelo REGULARIZADO foi menor, o que prova que o modelo REGULARIZADO tem maior capacidade de generaliza√ß√£o.

````python
# criando uma lista com as vari√°veis categ√≥ricas candidatas a DUMMIES
categoricals = ['condition', 
                'num_floors', 
                'is_waterfront', 
                'c_num_bath', 
                'c_num_partial_bath', 
                'zip']
# Criando uma fun√ß√£o que cria vari√°veis dummies das colunas escolhidas
def one_hot(df, cols):
    for each in cols:
        dummies = pd.get_dummies(df[each], prefix=each, drop_first=True)
        df = pd.concat([df, dummies], axis=1)
    return df

# aplicando a fun√ß√£o no DataFrame e nas colunas selecionadas
df = one_hot(df,categoricals)
# dropando as colunas originais
df = df.drop(categoricals, axis=1)
# visualizando os dados
df.head()
````

## Rodando as regress√µes lineares com features polinomiais

Agroa vamos rodas os modelos de Machine Learning utilizando os pacotes de regress√£o de Lasso e Ridge para entender como estes algoritmos lidam com a regulariza√ß√£o de par√¢metros.

Primeiramente vamos substituir os dados ogirinais da vari√°vel alvo price pelo log_price, pois como vimos isso melhora a resposta por corre√ß√£o da assimetria a varia√ß√£o da vari√¢ncia.

````python
# importando as bibliotecas
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import make_scorer, mean_squared_error, r2_score
from sklearn import linear_model
from sklearn.model_selection import cross_val_score
````

### Separando os dados em treino e teste

````python
# fun√ß√£o para separar os dados em treino e teste
def dfSplit(df,ratio,y='log_price'):
    train, test = train_test_split(df, test_size = ratio, random_state=0)
    y_train = train[y]
    y_test = test[y]
    x_train = train.ix[:, train.columns != y]
    x_test = test.ix[:, test.columns != y]
    return x_train, y_train, x_test, y_test
````

### Calculando vari√°veis polinomiais 

Fun√ß√£o para expandir o polinomio das colunas explicativas ou Xs. Isso serve para que a fun√ß√£o possa entender comportamos n√£o lineares por outro lado. Se tivermos intera√ß√µes e polinomios sem necessidade, podemos causar overfiting, ou seja, superajustar nosso modelo.

````python
# fun√ß√£o que calcula vari√°veis polinomiais
def tPoly(df, degree=1):
    polynomial = PolynomialFeatures(degree=degree)
    return polynomial.fit_transform(df)
````

````python
pd.DataFrame(tPoly(df))
````

# Fun√ß√µes para rodar os modelos de Machine Learning

Fun√ß√£o que roda os modelos de machine learning, aplica o polinomio, separa os dados em treino e teste, exibe as m√©tricas e plota os resultados. 

Como o Cross Validation Score est√° sempre tentando minimizar o erro (mse), o resultado m√©dio minimizado sempre acaba negativo, por isso, no momento de extrair a raiz (rmse), √© necess√°rio o sinal negativo. Para mais informac'~oes acesse __[MSE is negative when returned by cross_val_score](https://github.com/scikit-learn/scikit-learn/issues/2439)__

````python
# fun√ß√£o que roda os modelos de Machine Learning e aplica as outras fun√ß√µes 
scorer = make_scorer(mean_squared_error, greater_is_better = False)
def testRegs(df, clf, degree=1, ratio=.2, y='log_price', metrics=[]):
    # aplicando a fun√ß√£o para separa√ß√£o dos dados
    x_train, y_train, x_test, y_test = dfSplit(df,ratio,y='log_price')
    # aplicando a fun√ß√£o de c√°lculo de polinomiais nos dados de treino e teste
    poly_x_train = tPoly(x_train,degree)
    poly_x_test = tPoly(x_test,degree)
    # ajustando o modelo selecionado
    clf.fit(poly_x_train,y_train)
    # fazendo predi√ß√µes
    y_hat = clf.predict(poly_x_test)
    # calculando a raiz quadrada do erro m√©dio quadr√°tico RMSE (negativo pois o cross_val_score minimiza o valor resultante)
    rmse = np.sqrt(-cross_val_score(clf, poly_x_train, y_train, scoring = scorer, cv = 10))
    # imprimindo o RMSE
    print("RMSE:", rmse.mean())
    # fazendo predi√ß√µes com dados de treino e teste
    y_train_pred = clf.predict(poly_x_train)
    y_test_pred = clf.predict(poly_x_test)

    # imprimindo os resultados
    print('R2: %.2f, Score: %.2f, Parameters: %i' % (r2_score(y_test, y_hat), 
                                                     clf.score(poly_x_train,y_train), 
                                                     clf.coef_.shape[0]))
    # imprimindo os resultados
    #print('R2: %.2f, Score: %.2f, Parameters: %i' % (r2_score(y_test, y_hat), 
    #                                                 clf.score(poly_x_train,y_train), 
    #                                                 clf.coef_.shape[0]))
    
    # plotando uma dispers√£o entre os dados preditos de treino e teste pelos erros
    # para dados de treino em azul e dados de teste em verde
    plt.scatter(y_train_pred, y_train_pred - y_train, c = "blue", marker = "s", label = "Training data")
    plt.scatter(y_test_pred, y_test_pred - y_test, c = "lightgreen", marker = "s", label = "Validation data")
    plt.xlabel("Predicted values")
    plt.ylabel("Residuals")
    plt.hlines(y = 0, xmin = 10.5, xmax = 13.5, color = "red")
    plt.show()
    # plotando uma dispers√£o entre os dados preditos de treino e teste pelos dados de treino e teste
    # para dados de treino em azul e dados de teste em verde
    plt.scatter(y_train_pred, y_train, c = "blue", marker = "s", label = "Training data")
    plt.scatter(y_test_pred, y_test, c = "lightgreen", marker = "s", label = "Validation data")
    plt.xlabel("Predicted values")
    plt.ylabel("Real values")
    plt.legend(loc = "upper left")
    plt.plot([10.5, 13.5], [10.5, 13.5], c = "red")
    plt.show()
````

### Rodando uma regress√£o linear com as fun√ß√µes definidas

````python
# instanciando o modelo com a regress√£o linear
ols_simple = linear_model.LinearRegression()
# aplicando a fun√ß√£o para rodar a regress√£o
testRegs(df,ols_simple,1)
````

### Definindo novas fun√ß√µes para as regress√µes

````python
# fun√ß√£o para imprimir as m√©tricas do modelo
def print_metrics(y_train, yhat_train,y_test, yhat_test):
    # imprimindo dados de treino
    print('\n-----Dados de Treino-----')
    print('RMSE - treino', np.sqrt(mean_squared_error(y_train, yhat_train)))
    print('R2 - treino', r2_score(y_train, yhat_train))
    # imprimindo dados de teste
    print('\n-----Dados de Teste-----')
    print('RMSE - test', np.sqrt(mean_squared_error(y_test, yhat_test)))
    print('R2 - test', r2_score(y_test, yhat_test))
    print('\n------------------------')
````

````python
# fun√ß√£o para rodar os modelos com fun√ß√µes polinomiais
def run_model_with_poly(dataframe, poly_n, modelo, target='log_price'):
    # instanciando a fun√ß√£o polinomial    
    poly = PolynomialFeatures(degree=poly_n)
    # definindo oss dados e entrada
    X = df.drop([target], axis=1)
    # definindo a vair√°vel target
    newY = dataframe[target]
    # criando um novo DataFrame com as vari√°veis polinomiais e imprimindo o novo shape
    df_temp = pd.DataFrame(poly.fit_transform(X))
    print('\ndf_temp shape', df_temp.shape)
    # separando os dados em treino e teste
    X_train, X_test, y_train, y_test = train_test_split(df_temp,newY,test_size=0.2,random_state=0)
    # imprimindo os shapes dos dados de treino e teste
    print('\ntrain shape', X_train.shape)
    print('teste shape', X_test.shape)
    # ajustando o modelo
    modelo.fit(X_train, y_train)
    # fazendo predi√ß√µes para treino e teste
    yhat = modelo.predict(X_train)
    yhat_test = modelo.predict(X_test[list(X_train)])
    # imprimindo as m√©tricas
    print_metrics(y_train, yhat, y_test, yhat_test)
    return modelo
````

Dessa forma, o modelo de Regulariza√ß√£o a ser adotado deve ser o Lasso (L1).
